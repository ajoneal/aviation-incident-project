```python
import pandas as pd
import os
```

```python
path = os.listdir('./dataset/') # getting path
df = pd.DataFrame(list())
df.to_csv('./dataset/All.csv') # create an empty csv
```

```python
for i in path:
    if i != 'All.csv': # make sure it does not loop back to All.csv
        df = pd.read_csv('./dataset/All.csv') # read in All.csv
        
        df_temp = pd.read_json ('./dataset/'+i) # read in 'i' json file
        
        df_temp = df_temp.explode('cm_vehicles') # explode/tranform to a list
        df_x = (pd.DataFrame(df_temp['cm_vehicles'].apply(pd.Series))) # create list into its own DF
        
        # append new data from df_x
        df_temp['make'] = df_x['make']
        df_temp['model'] = df_x['model']
        df_temp['registrationNumber'] = df_x['registrationNumber']
        df_temp['numberOfEngines'] = df_x['numberOfEngines']
        df_temp['operatorName'] = df_x['operatorName']
        df_temp['registeredOwner'] = df_x['registeredOwner']
        df_temp['flightOperationType'] = df_x['flightOperationType']
        df_temp['flightScheduledType'] = df_x['flightScheduledType']
        df_temp['revenueSightseeing'] = df_x['revenueSightseeing'] 
        df_temp['secondPilotPresent'] = df_x['secondPilotPresent']
        
        df_temp.drop(columns='cm_vehicles', inplace=True) # drop column

        pd.concat([df, df_temp]).to_csv('./dataset/All.csv', index=False) # concat to All.csv
```

```python
df = pd.read_csv('./dataset/All.csv') # read back in the full All.csv
df.drop(columns='Unnamed: 0', inplace=True)
```
